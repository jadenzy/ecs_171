```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

np.random.seed(0)
X = 4 * np.random.rand(100, 1)
y = 1+ X + X**2 + X**3 + np.random.randn(100, 1)  # Adding noise to the output

#print("printing Original Data X:", X[:5])

# Create polynomial features
poly = PolynomialFeatures(degree=1)
X_poly = poly.fit_transform(X)

# print("printing Data X after transforming to polynomial features:", X_poly[:5])

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)


# Train polynomial regression model
model = LinearRegression()
model.fit(X_train, y_train)

print("Model weights: ")
print(model.coef_)


# Visualize the fit of the model on training data
plt.scatter(X_train[:,1], y_train, color='blue', label='Training Data')  # Plot original data points
plt.scatter(X_train[:,1], model.predict(X_train), color='red', label='Model Prediction')  # Plot the model prediction
plt.title('Model Fit on Training Data')
plt.xlabel('X_training')
plt.ylabel('Output')
plt.legend()
plt.show()



#Computing Bias and Variance 
import numpy as np
bias_test = np.mean((np.mean(model.predict(X_test))- y_test) ** 2)
print("Bias on Test Data:" , bias_test)
variance_test = np.mean((np.mean(model.predict(X_test))-model.predict(X_test))**2)
print("Variance on Test Data:" , variance_test)
```
